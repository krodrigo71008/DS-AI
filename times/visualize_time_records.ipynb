### World Model

```python
world_model_df = pd.read_csv("world_model.csv")
world_model_df["time"] /= 1000000 # converting from ns to ms
world_model_times = {k: v for (k, v) in world_model_df.groupby('type')}
```

```python
for name, time_df in world_model_times.items():
    print(name)
    print(time_df.describe())
    print("\n\n")
```

```python
for name, time_df in world_model_times.items():
    plt.figure()
    axes = time_df["time"].hist()
    axes.set_title(name)
```
